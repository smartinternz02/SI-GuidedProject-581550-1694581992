```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_crashes_data = sns.load_dataset('car_crashes')
```

```python
car_crashes_data
```

```python
# Vertical Bar Plot (Total Crashes by State)
plt.figure(figsize=(12, 6))
sns.barplot(x='abbrev', y='total', data=car_crashes_data, palette='viridis')
plt.title('Total Crashes by State (Vertical Bar Plot)')
plt.xlabel('State Abbreviation')
plt.ylabel('Total Crashes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie Chart (Proportion of Crashes by Distracted vs. Not Distracted)
distracted_crashes = car_crashes_data['total'] - car_crashes_data['not_distracted']
labels = ['Distracted', 'Not Distracted']
sizes = [distracted_crashes.sum(), car_crashes_data['not_distracted'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['coral', 'lightblue'])
plt.title('Proportion of Crashes by Distracted vs. Not Distracted (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
# Scatter Plot (Speeding vs. Alcohol Crashes)
sns.scatterplot(x='speeding', y='alcohol', data=car_crashes_data)
plt.title('Speeding vs. Alcohol Crashes (Scatter Plot)')
plt.xlabel('Speeding Crashes')
plt.ylabel('Alcohol Crashes')
plt.show()
```

```python
# Box Plot (Distribution of Insured Premium)
plt.figure(figsize=(8, 6))
sns.boxplot(x='ins_premium', data=car_crashes_data, color='skyblue')
plt.title('Distribution of Insured Premium (Box Plot)')
plt.ylabel('Insured Premium')
plt.show()
```

```python
# Histogram (Distribution of No Previous Accidents)
sns.histplot(car_crashes_data['no_previous'], bins=10)
plt.title('Distribution of No Previous Accidents (Histogram)')
plt.xlabel('Number of Accidents')
plt.ylabel('Frequency')
plt.show()
```

```python
# Violin Plot (Distribution of Losses Incurred by Insurance Companies)
plt.figure(figsize=(8, 6))
sns.violinplot(y='ins_losses', data=car_crashes_data, color='salmon')
plt.title('Distribution of Losses Incurred by Insurance Companies (Violin Plot)')
plt.ylabel('Losses Incurred')
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = car_crashes_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
